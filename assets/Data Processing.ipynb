```python
import pandas as pd
import datetime as dt
import numpy as np
```

```python
cities_data = pd.read_csv("cities.csv")
cities_data.head()
```

```python
cities_data["Date"] = cities_data["Date"].apply(lambda x:
                                               dt.datetime.fromtimestamp(x)
                                               )
cities_data["City"] = cities_data["City"].apply(lambda x:
                                               x.title())
del cities_data["City_ID"]
cities_data.head()
```

```python
html = cities_data.to_html(index=False)
print(html)
cities_data.to_html(open('asHtmlFile.html'), 'w')
save(html, 'F:\Bootcamp\Homework9\Web-Design-Challenge\tableHTML.html')
```
